```python
#Data Wrangling
```

```python
titanic.columns
```

```python
titanic2 = titanic.drop(['Unnamed: 0', 'adult_male', 'Unnamed: 16', 'fare2', 'embarked'], axis=1)
titanic2.head()
```

```python
#recode
```

```python
def sex (series):
    if series == "male":
        return 0
    if series == "female":
        return 1
titanic2['sex'] = titanic['sex'].apply(sex)
titanic2.head()
```

```python
def who (series):
    if series == "man":
        return 0
    if series == "woman":
        return 1
    if series == "child":
        return 2
titanic2['who'] = titanic['who'].apply(who)
titanic2.head()
```

```python
def alive (series):
    if series == "yes":
        return 0
    if series == "no":
        return 1
titanic2['alive'] = titanic['alive'].apply(alive)
titanic2.head()
```

```python
titanic2.info()
```

```python
titanic2['fare'] = titanic2['fare'].astype(int)
titanic2.head()
```

```python
titanic2.info()
```

```python
#Rename Columns
```

```python
titanic2.columns
```

```python
titanic2.rename(columns={'survived' : 'Survived'}, inplace=True)
titanic2.rename(columns={'pclass' : 'Pclass'}, inplace=True)
titanic2.rename(columns={'sex' : 'Sex'}, inplace=True)
titanic2.rename(columns={'age' : 'Age'}, inplace=True)
titanic2.rename(columns={'sibsp' : 'Siblings'}, inplace=True)
titanic2.rename(columns={'parch' : 'Parch'}, inplace=True)
titanic2.rename(columns={'fare' : 'Fare'}, inplace=True)
titanic2.rename(columns={'class' : 'Class'}, inplace=True)
titanic2.rename(columns={'who' : 'Gender'}, inplace=True)
titanic2.rename(columns={'deck' : 'Boat Deck'}, inplace=True)
titanic2.rename(columns={'embark_town' : 'Embark Town'}, inplace=True)
titanic2.rename(columns={'alive' : 'Alive'}, inplace=True)
titanic2.rename(columns={'alone' : 'Alone'}, inplace=True)
titanic2.head()
```
